### Rename the bankrupt? columns that indicate whether is True or False into "Statuts" 

```python
data.rename(columns={"Bankrupt?" : "Statuts"}, inplace = True )
data
```

```python
def clean_col_names(txt):
    txt = (
        txt.strip()
        .replace("?", "_")
        .replace("(", "_")
        .replace(")", "_")
        .replace(" ", "_")
        .replace("/", "_")
        .replace("-", "_")
        .replace("%", "")
        .replace("¥", "")
        .replace("__", "_")
        .replace("'", "")
        .lower()
    )
    return txt
```

```python
cols = list(data.columns)
cols = [clean_col_names(txt) for txt in cols]
data.columns = cols
```

We are going to visualize the correlation matrix of the dataset with a headmap. For that we use the Spearman correlation that measure monotonic relationship between feature and less sensitive to outliers than the well know pearson method(in case we have some in the data). we used adiverging color palette, with zero at the center, making it easy to identify strong positive and negative correlations between features.
Thanks to the headmap color we can easily identify the correlations areas (high or low).

```python
import matplotlib.pyplot as plt  # For plotting
import seaborn as sns            # For heatmap and color palettes
```

```python
f, ax = plt.subplots(figsize=(30, 25))
mat = data.corr("spearman") #calculates the Spearman correlation matrix 
mask = np.triu(np.ones_like(mat, dtype=bool)) #creates a mask that hides the upper triangle of the correlation matrix
cmap = sns.diverging_palette(230, 20, as_cmap=True) #creates a diverging color palette (show positive and neg correlation)
sns.heatmap(
    mat,
    mask=mask,  
    cmap=cmap,
    vmax=1,
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5}, #shrink the color bar size
)
plt.show()
```

From the map above we can identify the features that are positively/negatively correlate with that target. The color palrtte help to have an idea of the degreee of correlation. Many of the features are actually negatively correlate with the target. 

## Features selection
In the initial dataset we have 96 features ; the correlation matrix shows area of high and low correlation. We can imagine coirrelation between feature as well. We need to reduce the number of feature to include in the model in order to prevent multicolinearity issues. We can either rely on the cprellation approach to eliminarte some odf the feature or we can use a dimension reduction tools. Many tools are available when it commes to deal with high dimentionnality. Penalize methods like Ridge or Lasso migh be use , Princial component Analysis (PCA) me also be used.
Let at first use Lasso that will shrink non relevant coeficient toward zero in order to eliminate them and select relavant one.

```python
from sklearn.preprocessing import StandardScaler

features_df = data.drop(columns=['Statuts'])
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features_df)
y_lasso = data["Statuts"]
```

```python
from sklearn.linear_model import Lasso, LassoCV

# Lasso with cross-validation to automatically select the best alpha
lasso = LassoCV(cv=5, random_state=0)  # cv=5 for 5-fold cross-validation
lasso.fit(features_df, y_lasso)
```

```python
# Get selected features
selected_features = features_df.columns[(lasso.coef_ != 0)]
print("Selected Features:", selected_features.tolist())

# Filter your dataset to use only the selected features
features_selected = features_df[selected_features]
```

```python
features_df[selected_features]
```

The Lasso has selected 8 features that are the most relevant  out of 96 

Let's then using an simple correlation approch by taking advantage on the corrlation between the fearures and the target. 
Next we are going to identify the features with the highest positive and negative correlations with the target variable (status). So we are going to extract the feature that are highly correlated with the target and then map the correlation between them; visualizes their pairwise correlations in a heatmap. This helps to identify the relationships between the most predictive features and detect any multicollinearity.

```python
numeric_features = data.dtypes[data.dtypes != "int64"].index  # identify numeric variables (only float)

# Here search for higher positive correlation with the target
positive_corr = (
    data[numeric_features]
    .corrwith(data["Statuts"])
    .sort_values(ascending=False)[:10]
    .index.tolist()
)

# Similary to the previous, here search for higher negative correlation with the target 
negative_corr = (
    data[numeric_features].corrwith(data["Statuts"]).sort_values()[:10].index.tolist()
)

#Gardered positive and negative correlated data in a single dataframe
positive_corr = data[positive_corr + ["Statuts"]].copy()
negative_corr = data[negative_corr + ["Statuts"]].copy()
check_relation = (
    positive_corr.columns.tolist()[:-1] + negative_corr.columns.tolist()[:-1]
)
```

```python
mat = data[check_relation].corr("spearman")
mask = np.triu(np.ones_like(mat, dtype=bool))
plt.figure(figsize=(12, 11))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(
    mat,
    mask=mask,
    cmap=cmap,
    vmax=1,
    center=0,
    square=True,
    annot=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.5},
)
plt.title("Highest Corralated Features with Target", fontsize=18)

plt.show()
```

The map help to see the correlation between the features; we can then eliminate some of them in order to prevent multicollinearity issues.
